## Irreduzibel

Satz 4.6 Eine Matrix $A \in \mathbb{K}^{N, N}$ ist genau dann irreduzibel, wenn der zugeordnete Graph zusammenhängend ist.

Unter dem einer Matrix $A \in \mathbb{K}^{N, N}$ zugeordneten gerichteten Graphen $\mathcal{G}(A)$ versteht man folgende Konstruktion:

$\mathcal{G}(\boldsymbol{A})$ besteht aus $N$ Knoten $P_{1}, \ldots, P_{N}$
Eine gerichtete Kante $P_{i} \mapsto P_{j}$ verbindet $P_{i}$ und $P_{j}$ genau dann, wenn $a_{i j} \neq 0$ ist. (Für $a_{i i} \neq 0$ wird also $P_{i} \mapsto P_{i}$ mit sich selbst verbunden.)

Gerichtete Wege in $\mathcal{G}(A)$ bestehen aus aufeinanderfolgenden gerichteten Kanten. Der gerichtete Graph $\mathcal{G}(\boldsymbol{A})$ heißt zusammenhängend, falls für jedes Knotenpaar $\left(P_{i}, P_{j}\right), 1 \leq$ $i, j \leq N, i \neq j$ ein gerichteter Weg von $P_{i}$ nach $P_{j}$ besteht.

```julia
using Graphs, Plots, GraphRecipes
```

```julia
function irreduzibel(A::Matrix{Int64})::Bool
    g = DiGraph(A)
    display(graphplot(g, names=1:size(A)[1], curvature_scalar=0.1))
    return is_strongly_connected(g)
end
```

```julia
A = [7 3 4 0;
     0 5 0 3;
     3 1 -5 1;
     0 3 0 3]

println("A is irreducible: $(irreduzibel(A))")
```

```julia
B = [5 3 -1 0;
     -3 7 3 2;
     -2 2 5 -3;
     0 -1 1 5]

println("B is irreducible: $(irreduzibel(B))")
```
